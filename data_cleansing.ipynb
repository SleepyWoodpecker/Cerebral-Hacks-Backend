```python
import pandas as pd
```

```python
customer_df = pd.read_csv("data/customers.csv")
customer_df['user id'] = customer_df['Customer ID']
customer_df = customer_df[[
    'user id', 'Age', 'Gender', 'Item Purchased', 'Category',
    'Location', 'Size', 'Color', 'Season', 'Review Rating', 'Frequency of Purchases', 'Season']]
customer_df
```

```python
sales_df = pd.read_csv("data/sales.csv")


sales_df = sales_df[['user id', 'product id', 'Interaction type']].dropna()

sales_df
```

```python
products_df = pd.read_csv("data/products.csv")

products_df['product id'] = products_df['Uniqe Id']

products_df = products_df[["product id", "Product Name", "Category", "About Product", "Selling Price"]]

products_df
```

```python
df = pd.concat([customer_df, products_df, sales_df], axis=1).dropna()
df = df.loc[:, ~df.columns.duplicated()]
df[:100].to_json('data/data.json')
df
```
