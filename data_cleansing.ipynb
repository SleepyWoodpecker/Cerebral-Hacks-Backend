```python
import pandas as pd
```

```python
customer_df = pd.read_csv("data/customers.csv")
customer_df['user id'] = customer_df['Customer ID']
customer_df = customer_df[[
    'user id', 'Age', 'Gender', 'Item Purchased', 'Category',
    'Location', 'Size', 'Color', 'Season', 'Review Rating', 'Frequency of Purchases']][:100]
customer_df.to_json('data/customers.json')
customer_df
```

```python
sales_df = pd.read_csv("data/sales.csv")


sales_df = sales_df[['user id', 'product id', 'Interaction type']].dropna()[:100]

sales_df.to_json('data/sales.json')
```

```python
products_df = pd.read_csv("data/products.csv")

products_df['product id'] = products_df['Uniqe Id']

products_df = products_df[["product id", "Product Name", "Category", "Selling Price"]][:100]

products_df.to_json('data/products.json')
```

```python
sales_df
```

```python
df = pd.concat([customer_df, products_df, sales_df], axis=1)
df = df.loc[:, ~df.columns.duplicated()]
df.to_json('data/data.json')
df
```
